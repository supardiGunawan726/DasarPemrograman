<h1>Commenting</h1>

Komentar pada python dituliskan dengan menggunakan awalan tanda <b>#</b>

</ol>

Terdapat 3 macam bentuk komentar, yakni:

1.Single Line - (#) - digunakan untuk menjelaskan arti sebuah proses <br>

3.Multi Line - (""") - digunakan apabila memiliki isi/konten yang banyak<br>
