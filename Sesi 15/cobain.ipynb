```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataFrame = pd.read_csv("./data_mobil.csv")
dataFrame.head()
```

```python
dataFrame.groupby('tahun')[['harga_juta']].mean().plot()
plt.xlabel("Tahun kendaraan")
plt.ylabel("Harga (juta)")
plt.show()
```
